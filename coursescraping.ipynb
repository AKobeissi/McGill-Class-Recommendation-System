```python
def clean_professor_name(name):
    # Remove special characters and accents
    cleaned_name = unidecode.unidecode(name)
    # Remove middle names (assumes middle names are a single word)
    cleaned_name = ' '.join(cleaned_name.split()[:2])
    return cleaned_name
```
